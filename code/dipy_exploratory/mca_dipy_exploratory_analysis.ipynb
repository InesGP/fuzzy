```python
import numpy as np
import pandas as pd

import plotly.offline as off
import plotly.graph_objects as go
import plotly.express as px
```

```python
# Specify data file and load both dataframes from it
datafile = '/data/RocklandSample/derivatives/mca/mca.h5'
df = pd.read_hdf(datafile, 'graphs')
```

```python
sub = df['sub'].value_counts().to_dict()
df_ = df.query('noise_type == noise_type')
_df = df.query('noise_type != noise_type')
```

```python
for ss in sub:
    ref = _df.query('sub == "{0}"'.format(ss)).iloc[0]
    idx = df_.where(df['sub'] == ss).index
    df_.loc[idx, 'mse (percent difference)'] = 100 * df_.loc[idx, 'mse (ref)'] / ref['mse (self)']
```

```python
fig = px.violin(df_, y='mse (percent difference)', x='sub', color='noise_type',
                violinmode='overlay', points='all', box=False,
                hover_data=['simulation_id'])
fig.show()
```
